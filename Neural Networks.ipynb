```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data=pd.read_csv("C:\\Users\\dheni\\Downloads\\DS Assignments\\18.Neural networks\\Alphabets_data.csv")
data
```

```python
# Summarize key features
print("Number of samples:", data.shape[0])
print("Number of features:", data.shape[1] - 1)  # Excluding the letter column
print("Classes:", data['letter'].unique())
```

```python
# Import libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
# Missing values
data.isnull().sum()
```

```python
# Normalize features
x = data.drop('letter', axis=1)
y = data['letter']
```

```python
# Encode the labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_encoded
```

```python
# Scaling the data
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
# Split the dataset into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_encoded, test_size=0.2, random_state=42)
```

```python
# Splitting size
print(x_train.shape)
print(x_test.shape)
```

```python
# Import libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Import library
import warnings
warnings.filterwarnings('ignore')
```

```python
# Construct a basic ANN model
model = keras.Sequential()
model.add(layers.InputLayer(input_shape=(x_train.shape[1],)))
```

```python
# Adding one hidden layer with 32 neurons and ReLU activation
model.add(layers.Dense(32, activation='relu'))
```

```python
# Output layer for multi-class classification
model.add(layers.Dense(len(np.unique(y_encoded)), activation='softmax'))  # Assuming classification of multiple alphabets
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(x_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# Make predictions on the test set
y_pred = model.predict(x_test)
y_pred
```

```python
# Make Predictions on classes
y_pred_classes = np.argmax(y_pred, axis=1)
y_pred_classes
```

```python
# Import library
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Classification report
report = classification_report(y_test, y_pred_classes)
print(report)
```

I used Randomforestclassifier due to KerasClassifier is not supporting and it is showing module not found errors even after installing different versions of tensorflow and keras. So I used different classifier which can satisy the tasks.

```python
# Import libraries
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
# Create the Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42)
# Train the model
rf_classifier.fit(x_train, y_train)
```

```python
# Make predictions on the test set
y_pred = rf_classifier.predict(x_test)
y_pred
```

```python
# Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

```python
# Define the hyperparameter grid
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]}
```

```python
# Create a Random Forest model for hyperparameter tuning
rf_model = RandomForestClassifier(random_state=42)
rf_model
```

```python
# Perform grid search
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, scoring='accuracy', cv=3)
grid_search
```

```python
# Grid result
grid_result = grid_search.fit(x_train, y_train)
grid_result
```

```python
# Summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
# Get the best model
best_model = grid_result.best_estimator_
best_model
```

```python
# Make predictions with the best model
y_pred_best = best_model.predict(x_test)
y_pred_best
```

```python
# Evaluate the best model
print("Best Model Classification Report:")
print(classification_report(y_test, y_pred_best))
print(f"Best Model Accuracy: {accuracy_score(y_test, y_pred_best)}")
```
